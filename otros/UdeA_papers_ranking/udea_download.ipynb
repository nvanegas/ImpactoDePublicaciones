```python
import requests
from IPython.display import display, HTML, JSON
from pymongo import MongoClient
import time
import pandas as pd
import numpy as np
```

```python
db = MongoClient()["openalex_udea"]
col = db["works"]
col.drop()
```

```python
url="https://api.openalex.org/works?filter=authorships.institutions.lineage:I35961687,publication_year:2013-2023&per-page=100&cursor={}"
```

```python
done = False
next_cursor = "*"
counter = 0
while next_cursor :
    req=requests.get(url.format(next_cursor))
    if req.status_code == 200:
        data = req.json()
        if len(data["results"]) ==0:
            break
        for i in range(len(data["results"])):
            del data["results"][i]["abstract_inverted_index"]
        col.insert_many(data["results"])
        next_cursor=data['meta']['next_cursor']
        counter+=1
        if counter%10 == 0:
            print(next_cursor)
            print(f"{counter*100} de {data['meta']['count']}")
        time.sleep(0.1)
    else:
        print(req)
        break
        
```

```python
db = MongoClient()["scimago"]
col = db["stage"]
col.drop()
```

```python
for year in range(2013,2024):
    data = pd.read_csv(f"./scimago_{year}.csv",sep=";")
    records=[]
    for i in data[["SJR","H index","Issn","SJR Best Quartile","Title","Sourceid"]].index:
        rec={"year":year}
        rec["H index"] = int(data["H index"][i])
        if isinstance(data["SJR"][i],float):
            rec["SJR"] = data["SJR"][i]
        elif isinstance(data["SJR"][i],str):
            rec["SJR"] = float(data["SJR"][i].replace(",","."))

        if np.isnan(rec["SJR"]):
            #continue
            rec["SJR"] = -1  # si es nan que es lo que nos salia, dejo en -1 
        issn=[]
        for value in data["Issn"][i].split(","):
            issn.append(value.strip())
        rec["Issn"] = issn 
        rec["H index"] = int(data["H index"][i])
        rec["SJR Best Quartile"] = data["SJR Best Quartile"][i]
        rec["Title"] = data["Title"][i]
        rec["Sourceid"] = str(data["Sourceid"][i])
        records.append(rec)
    col.insert_many(records)
```
